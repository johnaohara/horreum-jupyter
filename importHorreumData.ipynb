```python
import requests
import matplotlib.pyplot as plt
import json
import pandas as pd


url = "https://example.com/api/grafana/query"
payload = '{"range":{"from":"2022-12-20T15:23:53.984Z","to":"2023-01-20T15:23:53.984Z","oneBeforeAndAfter":true},"targets":[{"target":"119351;","type":"timeseries","refId":"ignored"},{"target":"119355;","type":"timeseries","refId":"ignored"},{"target":"119352;","type":"timeseries","refId":"ignored"},{"target":"119356;","type":"timeseries","refId":"ignored"},{"target":"119353;","type":"timeseries","refId":"ignored"},{"target":"119354;","type":"timeseries","refId":"ignored"}]}'
headers = {
    "Content-Type": "application/json",
    "Accept": "application/json",
    "Origin": "https://example.com",
    "Referer": "https://example.com/changes?test=Quarkus%20-%20getting-started%20-%20JVM",
    "cache-control": "no-cache",
    "cookie": "",
}

response = requests.request("POST", url, data=payload, headers=headers)
#convert string to  object

dataset = json.loads(response.text)
```
